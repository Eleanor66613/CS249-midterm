##Population Dynamics:  Competition in a World of Memes

The life-cycle curves in news articles track the <b>population size</b> of the articles.
At first the population shoots up, then levels off, and starts to drop steeply.
Other news memes fight for attention, and older memes lose their staying power.
Eventually all instances of the older meme can disappear.

You have heard of "Predator-Prey models'. 
The field of Population Dynamics (http://en.wikipedia.org/wiki/Population_dynamics) develops mathematical models of processes like this.  The curve above resembles that of a prey that initially multiplies rapidly but is then overcome by predators.
A popular mathematical formulation using differential equations is known as the Lotka-Volterra equation
(http://en.wikipedia.org/wiki/Lotka%E2%80%93Volterra_equation).
Predator-prey models measure the effect of resources (like food, or in new cycles, like attention of readers) on population sizes.
Limited resources cause competition, which can then lead to large increases and decreases in population size.

The point here is that the curve shape above might somehow be modeled by a curve that arises in population dynamics,
and there might actually be a justifiable mathematical model for these curves.
This is the point of departure for this problem:  is there a model for the life-cycle of baby names?

# Getting the "Baby Memes" Data

```python
%%R
## the sorted_names.csv was attained by a C++ program using hash table for O(n) time
bnames <- read.csv("sorted_names.csv")
n <- dim(bnames)[1]
p <- dim(bnames)[2]
print(dim(bnames))
print(head(bnames))
```

```python
%%R
year <- seq(1880,2012)
alex <- bnames[bnames$name =='Alex' & bnames$sex=='M',]
plot(year, alex[,3:(p-1)], type='l')
```

```python
%%R -w 1000 -h 600
########
library(MASS)
library(ggplot2)
library(splines)
library(gridExtra)

# test
```

```python
%%R -w 1000 -h 600
# choose 100 most frequent names
# and draw the fitted curve
library(reshape)
plots <- list()  # new empty list
for(i in 1:20){
    birth_per <- bnames[i, 3:(p-1)]/bnames[i,p]
    cur_data <- rbind(year, birth_per)
    cur_data <- as.data.frame(t(cur_data))
    colnames(cur_data) <- c("year", "birth_per")
    model <- lm(birth_per ~ ns(year, df=6), data=cur_data)
    birth_hat <- predict(model)
    
    cur_data$birth_hat <- birth_hat
    df <- melt(data=cur_data, id.vars='year')
    
    #print(summary(model)$adj.r.squared)

    pp <- ggplot(data=df, aes(x=year, y=value, colour=variable)) + 
        geom_line() + ggtitle(paste(bnames[i,1], round(summary(model)$adj.r.squared, 2)))
    plots[[i]] <- pp
}
args.list <- c(plots, list(nrow=5,ncol=4))

do.call(grid.arrange, args.list)

## Following is a graph of the fitted curves versus the original curve.
## The red the is original graph, and the blue one is the fitted curve
## The two digits number following the name in the title is the adjusted R squre.
```

```python
%%R
# Now we do the fitting for all the names, 
# and filter out the fitted curves with R^2 less than 0.90

# note n= 101261, it may be time-comsuming to compute all the names
# first we'll choose the first 10,000
rs_c <- 0.85
topnum <-0
## for matched curves, we want to extract the time of occurance along with its rank
ocur_num <- NULL
name_rank <- NULL
for(i in 1:n){
    birth_per <- bnames[i, 3:(p-1)]/bnames[i,p]
    cur_data <- rbind(year, birth_per)
    cur_data <- as.data.frame(t(cur_data))
    colnames(cur_data) <- c("year", "birth_per")
    model <- lm(birth_per ~ ns(year, df=6), data=cur_data)
    if(summary(model)$adj.r.squared > rs_c){
        topnum <- topnum +1
        ## extract the time of occurance along with its rank
        ocur_num <- c(ocur_num, bnames[i, p])
        name_rank <- c(name_rank, topnum)
        ## extract the top 100 names to plot
        if( topnum < 100){
            birth_hat <- predict(model)
            cur_data$birth_hat <- birth_hat
            df <- melt(data=cur_data, id.vars='year')
            name <- paste("top_",topnum, sep="")
            assign(name, df)
            name <- paste("top_",topnum,"_Rs", sep="")
            assign(name, summary(model)$adj.r.squared )
        }
    }
}
```

```python
%%R
print("Number of reliable matches")
print(length(ocur_num))
print("Percentage over all the names")
print(length(ocur_num)/n) 

## Comment:
## The following percentage can be further improved by increase the critical R square value
## Currently we set the R square value to be greater than 0.9 to determine a "good fit"
## If we make the bound looser the percentage will be higher
```

```python
%%R
ocur_data <- as.data.frame(cbind(ocur_num, name_rank))
print(summary(ocur_data))
plot(log(ocur_data$ocur_num), log(ocur_data$name_rank), cex=0.4)
plot(ocur_num, name_rank, cex=0.4)
```

```python
%%R -w 1000 -h 600
# choose 100 most frequent names
# and draw the fitted curve
library(reshape)
plots <- list()  # new empty list
for(i in 1:20){
    name <- paste("top_", i, sep="")
    Rname <- paste("top_",i,"_Rs", sep="")
    pp <- ggplot(data=get(name), aes(x=year, y=value, colour=variable)) + 
        geom_line() + ggtitle(paste(bnames[i,1], round(get(Rname), 2)))
    plots[[i]] <- pp
}
args.list <- c(plots, list(nrow=5,ncol=4))
do.call(grid.arrange, args.list)
```
